```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests

df = pd.read_excel("时间关系.xlsx")
grangercausalitytests(df[['PD', 'LSI']], maxlag=10, addconst=True, verbose=True)
```

```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests
import matplotlib.pyplot as plt
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")

result = grangercausalitytests(df[["PD","LPI"]], maxlag=10, addconst=True, verbose=True)
result_list = []
for lag in result:
    for test in result[lag][0]:
        lag_info = []
        lag_info["lag"] = lag
        lag_info["p_value"] = result[lag][0][test][1]
        lag_info["test_type"] = test
        result.append(lag_info)
#         result.update(lag_info)
test_result = pd.DataFrame(result_list)

fig = plt.figure(figsize=(12, 6))

for t in test_result['test_type'].unique():
    t_data = test_result[test_result['test_type'] == t]
    plt.plot(t_data['lag'],  # x轴数据
             t_data['p_value'],  # y轴数据
             linestyle='-',  # 折线类型
             linewidth=2,  # 折线宽度
             color='b',  # 折线颜色
             marker='o',  # 点的形状
             markersize=6,  # 点的大小
             markeredgecolor='r',  # 点的边框色
             markerfacecolor='r')  # 点的填充色

plt.axhline(y=0.05, color='c', linestyle='--')
plt.title('格兰杰因果关系检验结果')
plt.xlabel('lag')
plt.ylabel('p_value')
plt.show()
```
